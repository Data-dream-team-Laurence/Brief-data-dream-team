**Nous calculons la fréquence de commande d'un client.**

```python
frequence = pd.merge(customers , orders, on = 'customer_id' , how = 'inner')
frequence
```

```python
# converting date columns to datetime
date_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_columns:
    frequence[col] = pd.to_datetime(frequence[col], format='%Y-%m-%d %H:%M:%S')
```

```python
frequence.info()
```

```python
#frequence['frequence'] = frequence['customers_id'] frequence['order_purchase_timestamp']
df_frequency = frequence.groupby('customer_unique_id').order_purchase_timestamp.count().reset_index()
df_frequency.columns = ['customer_unique_id','order_purchase_timestamp']

#add this data to our main dataframe
df_frequency.describe()
```
